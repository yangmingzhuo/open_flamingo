```python
import json
import os
import matplotlib.pyplot as plt
import skimage.io as io
import requests
```

```python
RT_res = json.load(open('/data/wyl/open_flamingo/output_CLIP_new/cocoresults_CLIP_4.json', 'r'))
GT_res = json.load(open('gt.json', 'r'))

image_dir_path = '/data/wyl/coco_data/val2014'
def print_images(image_ids, r_texts):
    for i, id in enumerate(image_ids):
        try:
            I = io.imread(f"/data/wyl/coco_data/val2014/COCO_val2014_{id:012d}.jpg")
        except:
            I = io.imread(f"/data/wyl/coco_data/train2014/COCO_train2014_{id:012d}.jpg")
        plt.imshow(I)
        plt.axis('off')
        plt.show()
        print(id)
        print("RT:", r_texts[i])
```

```python
i =  0

I = io.imread(f"{image_dir_path}/COCO_val2014_{RT_res[i]['image_id']:012d}.jpg")
plt.imshow(I)
plt.axis('off')
plt.show()

print("GT:")
for c in GT_res[i]['captions']:
    print(c)

print()
print("RT:")
print(RT_res[i]['caption'])
# print(RT_res[i]['prompt_text'])

print_images(RT_res[i]["prompt_images"], RT_res[i]['prompt_texts'])

    
```
