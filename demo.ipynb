```python
import json
import os
import matplotlib.pyplot as plt
import skimage.io as io
import requests
```

```python
baseline_res = json.load(open('./output_baseline/cocoresults_baseline_4.json', 'r'))
RT_res = json.load(open('./output_RT/cocoresults_RT_4.json', 'r'))
GT_res = json.load(open('gt.json', 'r'))

image_dir_path = '/data/wyl/coco_data/train2014'
def print_images(image_ids, b_texts, r_texts):
    for i, id in enumerate(image_ids):
        I = io.imread(f"{image_dir_path}/COCO_train2014_{id:012d}.jpg")
        plt.imshow(I)
        plt.axis('off')
        plt.show()
        print(id)
        print("baseline:", b_texts[i])
        print("RT:", r_texts[i])
```

```python
i = 1

I = io.imread(f"{image_dir_path}/COCO_train2014_{baseline_res[i]['image_id']:012d}.jpg")
plt.imshow(I)
plt.axis('off')
plt.show()
print(baseline_res[i]['image_id'])

print("GT:")
for c in GT_res[i]['captions']:
    print(c)

print()
print("baseline:")
print(baseline_res[i]['caption'])
# print(baseline_res[i]['prompt_text'])

print()
print("RT:")
print(RT_res[i]['caption'])
# print(RT_res[i]['prompt_text'])

print_images(baseline_res[i]['prompt_images'], baseline_res[i]['prompt_texts'], RT_res[i]['prompt_texts'])

    
```
